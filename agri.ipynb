## Reduce

```python
def add(x,y):
    return x+y
total_price=reduce(add,df['Price'])
print("total price:",total_price)
```

```python
def max_value(x,y):
    return x if x > y else y
max_price=reduce(max_value,df['Price'])
print("maximam price:",max_price)
```
